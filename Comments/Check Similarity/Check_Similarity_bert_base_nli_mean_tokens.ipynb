```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
```

```python
model_name = 'bert-base-nli-mean-tokens'
```

```python
def get_data(data_file_path):
    df = pd.read_csv(data_file_path)

    return df['comment'].to_list(), df['user_gender'].to_list()
```

```python
comments, genders = get_data('/content/drive/MyDrive/Datasets/dataset.csv')
n = int(len(comments) / 12)
x = [comments[i:i + n] for i in range(0, len(comments), n)]
y = [genders[i:i + n] for i in range(0, len(genders), n)]

original_comments = x[0]
labels = y[0]

replaced_comments, replaced_labels = get_data('/content/drive/MyDrive/Datasets/Bert Modal On One Comment Full Data/1_replaced_dataset_one_comment.csv')
```

```python
model = SentenceTransformer(model_name)
```

```python
original_comments_vector = model.encode(original_comments)
```

```python
replaced_comments_vector = model.encode(replaced_comments)
```

```python
similarity_score = []

for i in tqdm(range(len(original_comments))):
    similarity_score.append(cosine_similarity([original_comments_vector[i]], [replaced_comments_vector[i]])[0][0])
```

```python
df = pd.DataFrame(list(zip(labels,original_comments, replaced_comments, similarity_score)), columns=['user_gender', 'original_comment', 'replaced_comment', 'similarity_score'])

df.to_csv('/content/drive/MyDrive/Datasets/Bert Modal On One Comment Full Data/1_similarity_dataset_one_comment.csv')
```
